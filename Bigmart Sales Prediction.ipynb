## Dataset Information

The data scientists at BigMart have collected 2013 sales data for 1559 products across 10 stores in different cities. Also, certain attributes of each product and store have been defined. The aim is to build a predictive model and find out the sales of each product at a particular store.

Using this model, BigMart will try to understand the properties of products and stores which play a key role in increasing sales.


Variable | Description
----------|--------------
Item_Identifier | Unique product ID
Item_Weight | Weight of product
Item_Fat_Content | Whether the product is low fat or not
Item_Visibility | The % of total display area of all products in a    store allocated to the particular product
Item_Type | The category to which the product belongs
Item_MRP | Maximum Retail Price (list price) of the product
Outlet_Identifier | Unique store ID
Outlet_Establishment_Year | The year in which store was established
Outlet_Size | The size of the store in terms of ground area covered
Outlet_Location_Type | The type of city in which the store is located
Outlet_Type | Whether the outlet is just a grocery store or some sort of supermarket
Item_Outlet_Sales | Sales of the product in the particulat store. This is the outcome variable to be predicted.

## Preprocessing the dataset

```python
# check for null values
df.isnull().sum()
```

```python
# check for categorical attributes
cat_col = []
for x in df.dtypes.index:
    if df.dtypes[x] == 'object':
        cat_col.append(x)
cat_col
```

```python
cat_col.remove('Item_Identifier')
cat_col.remove('Outlet_Identifier')
cat_col
```

```python
# print the categorical columns
for col in cat_col:
    print(col)
    print(df[col].value_counts())
    print()
```

```python
# fill the missing values
item_weight_mean = df.pivot_table(values = "Item_Weight", index = 'Item_Identifier')
item_weight_mean
```

```python
miss_bool = df['Item_Weight'].isnull()
miss_bool
```

```python
for i, item in enumerate(df['Item_Identifier']):
    if miss_bool[i]:
        if item in item_weight_mean:
            df['Item_Weight'][i] = item_weight_mean.loc[item]['Item_Weight']
        else:
            df['Item_Weight'][i] = np.mean(df['Item_Weight'])
```

```python
df['Item_Weight'].isnull().sum()
```

```python
outlet_size_mode = df.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
outlet_size_mode
```

```python
miss_bool = df['Outlet_Size'].isnull()
df.loc[miss_bool, 'Outlet_Size'] = df.loc[miss_bool, 'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
```

```python
df['Outlet_Size'].isnull().sum()
```

```python
sum(df['Item_Visibility']==0)
```

```python
# replace zeros with mean
df.loc[:, 'Item_Visibility'].replace([0], [df['Item_Visibility'].mean()], inplace=True)
```

```python
sum(df['Item_Visibility']==0)
```

```python
# combine item fat content
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'})
df['Item_Fat_Content'].value_counts()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Outlet'] = le.fit_transform(df['Outlet_Identifier'])
cat_col = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'New_Item_Type']
for col in cat_col:
    df[col] = le.fit_transform(df[col])
```

## Onehot Encoding

```python
df = pd.get_dummies(df, columns=['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'New_Item_Type'])
df.head()
```

## Input Split

```python
X = df.drop(columns=['Outlet_Establishment_Year', 'Item_Identifier', 'Outlet_Identifier', 'Item_Outlet_Sales'])
y = df['Item_Outlet_Sales']
```

## Model Training

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
def train(model, X, y):
    # train the model
    model.fit(X, y)
    
    # predict the training set
    pred = model.predict(X)
    
    # perform cross-validation
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE:",mean_squared_error(y,pred))
    print("CV Score:", cv_score)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  


model = LinearRegression()
model.fit(X_scaled, y)

coef = pd.Series(model.coef_, index=X.columns).sort_values()
coef.plot(kind='bar', title="Model Coefficients")
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalize X

# Train the model
model = Ridge()  # Remove normalize=True
model.fit(X_scaled, y)

# Get model coefficients
coef = pd.Series(model.coef_, index=X.columns).sort_values()
coef.plot(kind='bar', title="Model Coefficients")
```

```python
model = Lasso()
train(model, X, y)
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind='bar', title="Model Coefficients")
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
```

# Findings from the BigMart Sales Prediction Analysis
## 1. Data Overview & Preprocessing
The dataset contains 1559 products across 10 stores, with sales data from 2013.

There were missing values in Item_Weight and Outlet_Size, which were filled using:

Item Weight: Filled using the average weight of the same Item_Identifier.

Outlet Size: Imputed using the most frequent category within the same Outlet_Type.

Zero values in Item_Visibility were replaced with the mean of the column.

Inconsistent values in Item_Fat_Content were corrected by grouping similar values (e.g., ‘LF’ → ‘Low Fat’).

A new feature New_Item_Type was created by categorizing products into Food, Drinks, and Non-Consumable based on their identifiers.

## 2. Exploratory Data Analysis (EDA)
Sales Distribution:

The Item_Outlet_Sales variable was skewed, and a log transformation was applied to normalize it.

Feature Distributions:

Item_Weight, Item_Visibility, and Item_MRP had varying distributions, with some outliers.

Categorical Analysis:

Outlet_Establishment_Year showed that older stores have lower sales.

Outlet_Type revealed that Supermarket Type 3 had the highest sales, while grocery stores had the lowest.

## 3. Feature Engineering
Encoded categorical features using Label Encoding and One-Hot Encoding to convert them into numerical values for model training.

A new feature, Outlet_Years, was created to calculate the age of each store.

## 4. Model Training & Performance
Linear Models (Linear Regression, Ridge, Lasso)
Linear Regression & Ridge Regression:

Performed poorly due to multicollinearity and the inability to capture complex relationships.

Item_MRP had the highest impact on sales prediction.

Lasso Regression:

Applied feature selection by reducing the number of features with low importance.

Tree-Based Models (Decision Tree, Random Forest, Extra Trees)
Decision Tree Regressor:

Overfitted the data, leading to poor generalization.

Most important features: Item_MRP, Outlet_Type, Item_Visibility.

Random Forest Regressor:

Performed significantly better by reducing overfitting.

Most important features: Item_MRP, Outlet_Type, Outlet_Years.

Extra Trees Regressor:

Provided the best feature importance insights, with Item_MRP being the most significant predictor.

## 5. Key Insights & Business Implications
Item MRP is the strongest predictor of sales, suggesting that pricing strategy is critical for revenue optimization.

Outlet Type heavily influences sales – larger supermarkets perform better than smaller grocery stores.

Older stores tend to have lower sales, indicating the need for store upgrades or better marketing strategies.

Product visibility plays a role, so optimizing product placements could boost sales.

Non-food items behave differently than food items in terms of sales, implying different sales strategies for different product categories.

# Challenges Faced in the BigMart Sales Prediction Analysis
## 1. Data Quality Issues
Missing Values:

Item_Weight and Outlet_Size had missing data, requiring imputation, which may have introduced biases.

Zero Values in Item Visibility:

Some products had Item_Visibility = 0, which is unrealistic and required correction.

## 2. Business Challenges
Changing Market Trends:

Past sales data may not fully reflect future trends due to economic shifts, inflation, or changing consumer behavior.

Promotions and Seasonal Effects:

The dataset does not include promotional campaigns, holidays, or seasonal trends, which could impact sales.
